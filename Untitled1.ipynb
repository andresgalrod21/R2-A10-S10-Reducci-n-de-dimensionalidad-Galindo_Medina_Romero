```python
# Paso 1: Importar libreras necesarias
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Paso 2: Cargar el dataset
# Aqu asumimos que subiste el archivo "max_data.csv" en Colab
data = pd.read_csv('data.csv')

# Paso 3: Seleccionar las columnas necesarias para el anlisis de K-Means
# Vamos a utilizar 'imdbAverageRating' y 'imdbNumVotes' para realizar el clustering
selected_data = data[['imdbAverageRating', 'imdbNumVotes']].dropna()

# Paso 4: Escalar los datos para que el K-means funcione mejor
scaler = StandardScaler()
scaled_data = scaler.fit_transform(selected_data)

# Paso 5: Determinar el nmero ptimo de clusters usando el mtodo de Elbow
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Graficar el mtodo de Elbow
plt.figure(figsize=(8, 5))
plt.plot(range(1, 10), inertia, marker='o')
plt.xlabel('Nmero de clusters (k)')
plt.ylabel('Inercia')
plt.title('Mtodo de Elbow para determinar el nmero ptimo de clusters')
plt.show()

# Paso 6: Aplicar K-means con el nmero de clusters elegido (ejemplo con k=3)
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(scaled_data)

# Aadir el resultado del cluster al DataFrame original
selected_data['Cluster'] = clusters

# Paso 7: Visualizar los clusters resultantes
plt.figure(figsize=(10, 6))
plt.scatter(selected_data['imdbAverageRating'], selected_data['imdbNumVotes'], c=clusters, cmap='viridis')
plt.xlabel('IMDb Average Rating')
plt.ylabel('IMDb Num Votes')
plt.title('Clusters de ttulos en la plataforma MAX')
plt.colorbar(label='Cluster')
plt.show()

# Paso 8: Interpretacin de Resultados
# En este paso, pueden ver que los ttulos estn agrupados segn sus ratings y cantidad de votos.
# Esto puede ayudar a entender qu tipos de ttulos tienden a ser ms populares o tener mejores calificaciones.
```
